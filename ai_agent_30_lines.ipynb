## 🎬 2. AI Agent in 30 lines

Smallest possible CrewAI program: **1 Agent + 1 Task + 1 Crew**. No tools, no memory.

```python
from crewai import Agent, Task, Crew, Process

# 1. Define an Agent — role + goal + backstory is all you need
researcher = Agent(
    role="Senior AI Researcher",
    goal="Explain complex AI concepts in plain English with concrete examples",
    backstory=(
        "You teach AI to working developers. You hate buzzwords. "
        "You always lead with a concrete example before any theory."
    ),
    llm=MODEL,
    verbose=True,
)

# 2. Define a Task
task = Task(
    description="Explain what an AI Agent is in under 100 words. Lead with a real example.",
    expected_output="A 100-word plain-English explanation starting with a concrete example.",
    agent=researcher,
)

# 3. Run the Crew
crew = Crew(agents=[researcher], tasks=[task], process=Process.sequential, verbose=True)
result = crew.kickoff()

print("\n=== AGENT OUTPUT ===\n")
print(result.raw)
```

```python
# --- Tool 1: Free web search via DuckDuckGo (no API key!) ---
from crewai import Agent, Task, Crew
from crewai.tools import tool
from ddgs import DDGS


@tool("Web Search")
def web_search(query: str) -> str:
    """Search the web with DuckDuckGo. Returns the top 5 results as 'title — snippet (url)'.
    Use this whenever you need fresh, current information from the internet."""
    with DDGS() as ddgs:
        hits = list(ddgs.text(query, max_results=5))
    if not hits:
        return "No results."
    return "\n".join(
        f"- {h.get('title','')} — {h.get('body','')[:160]} ({h.get('href','')})"
        for h in hits
    )


web_agent = Agent(
    role="Tech News Scout",
    goal="Find the latest verified info on any AI topic",
    backstory="You search the web, read sources, and only report verified facts with links.",
    tools=[web_search],
    llm=MODEL,
    verbose=True,
)

task = Task(
    description="What is Anthropic's most recent model release? Include the date and one key feature.",
    expected_output="Model name, release date, and one feature — with a source URL.",
    agent=web_agent,
)

result = Crew(agents=[web_agent], tasks=[task], verbose=True).kickoff()
print("\n=== WEB AGENT OUTPUT ===\n")
print(result.raw)
```
