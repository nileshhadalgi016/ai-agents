```python
import os
from getpass import getpass

# Get a free key from https://aistudio.google.com/apikey
if not os.getenv("GEMINI_API_KEY"):
    os.environ["GEMINI_API_KEY"] = getpass("GEMINI_API_KEY: ")

# CrewAI uses LiteLLM, which also reads GOOGLE_API_KEY — mirror it.
os.environ["GOOGLE_API_KEY"] = os.environ["GEMINI_API_KEY"]

MODEL = "gemini/gemini-2.5-flash"
print(f"Using model: {MODEL}")
```

## 🎬 2. AI Agent in 30 lines

Smallest possible CrewAI program: **1 Agent + 1 Task + 1 Crew**. No tools, no memory.

```python
from crewai import Agent, Task, Crew, Process

# 1. Define an Agent — role + goal + backstory is all you need
researcher = Agent(
    role="Senior AI Researcher",
    goal="Explain complex AI concepts in plain English with concrete examples",
    backstory=(
        "You teach AI to working developers. You hate buzzwords. "
        "You always lead with a concrete example before any theory."
    ),
    llm=MODEL,
    verbose=True,
)

# 2. Define a Task
task = Task(
    description="Explain what an AI Agent is in under 100 words. Lead with a real example.",
    expected_output="A 100-word plain-English explanation starting with a concrete example.",
    agent=researcher,
)

# 3. Run the Crew
crew = Crew(agents=[researcher], tasks=[task], process=Process.sequential, verbose=True)
result = crew.kickoff()

print("\n=== AGENT OUTPUT ===\n")
print(result.raw)
```
